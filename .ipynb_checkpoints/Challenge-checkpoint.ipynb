```python
def partition_exponential_smoothing_model(chunk, param_list, 
                                          test_years=np.arange(2013, 2016), damped_trend=False):
    '''
        Calculate the simple additive exponential smoothing for each of the series, i.e.,
        groups of (country, indicator) and return the trained model for each partitition,
        together with some other informations used for value correction.
        The parameters of the model are tunned using the Corrected Akaike Information Criterion (AICC).
    '''
    train_series = chunk[['year', 'value']].sort_values(by='year').values
    future_series = train_series[train_series[:, 0] > test_years.max()]
    train_series = train_series[train_series[:, 0] < test_years.min()]
    year_min, year_max = train_series[:, 0].min(), train_series[:, 0].max()
    missing_years = np.setdiff1d(np.arange(year_min, year_max), train_series[:, 0])
    interp_arr = np.interp(missing_years, train_series[:, 0], train_series[:, 1])
    interp_arr = np.vstack([missing_years, interp_arr]).T
    train_series = np.vstack([train_series, interp_arr])
    train_series = train_series[np.argsort(train_series[:, 0])]
    
    ''' Time series with negative values 
    => don't accept seasonal periods nor multiplicative trend'''
    neg_value = False
    AIC_array = np.array([], dtype='float')
    if np.any(train_series[:, 1] <= 0.0):
        neg_value = True
        params = {'trend': 'add', 'seasonal': None}
        exp_smooth = ExponentialSmoothing(train_series[:, 1], **params, 
                                          damped=damped_trend).fit()
        best_params = params
            
    else:
        AIC_array = np.array([], dtype='float')
        for params in param_list:
            exp_smooth = ExponentialSmoothing(train_series[:, 1], **params,
                                             damped=damped_trend).fit()
            AIC_array = np.hstack([AIC_array, [exp_smooth.aicc]])
        
        min_idx = np.argmin(AIC_array)
        best_params = param_list[min_idx]
        exp_smooth = ExponentialSmoothing(train_series[:, 1], 
                                          **best_params,
                                         damped=damped_trend).fit()

    return {
        'model': exp_smooth,
        'first_year': train_series[:, 0].min(),
        'future_series': future_series,
        'best_params': best_params,
        'has_future': future_series.size > 0,
        'negative_value': neg_value
    }
```

```python
def apply_forecast_to_partition(chunk, map_funcs, alpha=1.0, testing=False):
    '''
        Apply the specific forecast model for each series (partition) and
        perform some corrections when there are future values for the predictions.
    '''
    chunk = chunk.sort_values(by='year')
    country, indicator = chunk[['country', 'indicator']].iloc[0].values
    if not testing:
        map_func = map_funcs.loc[pd.IndexSlice[country, indicator]]
    else:
        map_func = map_funcs
    pred_years = chunk.year.values
    first_year = map_func['first_year']
    start_idx = int(pred_years.min()-first_year)
    end_idx = int(pred_years.max()-first_year)
    preds = map_func['model'].predict(start=start_idx, end=end_idx)
    preds = np.vstack([np.arange(first_year+start_idx, first_year+end_idx+1), preds]).T
    preds = preds[np.isin(preds[:, 0], pred_years)][:, 1]
    if(map_func['has_future']):
        future_series = map_func['future_series']
        dist_weights = (
            1 / np.abs(pred_years[:, np.newaxis] 
                       - future_series[:, 0]))
        diff_values = preds[:, np.newaxis] - future_series[:, 1]
        preds = preds - alpha*np.sum(dist_weights*diff_values, axis=1)/np.sum(dist_weights)
    
    return pd.concat([chunk.reset_index(drop=True), 
                      pd.Series(preds, name='value')], axis=1)
```

```python
data_dd = dd.read_csv('data/data.csv', blocksize=10e6)
data_dd.head()
```

```python
param_list = [
        {'trend': 'add', 'seasonal': None},
        {'trend': 'mul', 'seasonal': None},
]
```

```python
model_obj = data_dd.groupby(['country', 'indicator'])\
        .apply(lambda x: partition_exponential_smoothing_model(x, param_list),
              meta=dict)
```

```python
model_obj.visualize()
```

```python
MODEL_PATH = 'model/model_map.pkl'
if not os.path.exists(MODEL_PATH):
    model_obj = model_obj.compute()
    gc.collect()
    with open(MODEL_PATH, 'wb') as f:
        pkl.dump(model_obj, f)
else:
    with open(MODEL_PATH, 'rb') as f:
        model_obj = pkl.load(f)
```

```python
model_obj
```

```python
test_df = pd.read_csv('data/test.csv')
test_df.head()
```

```python
test_df.info()
```

```python
RESULT_PATH = 'result/results.csv'
if not os.path.exists(RESULT_PATH):
    results = test_df\
              .reset_index()\
              .groupby(['country', 'indicator'])\
              .apply(lambda x: apply_forecast_to_partition(x, model_obj))
    results = results.reset_index(drop=True).set_index('index').sort_index()
    results.to_csv(RESULT_PATH, index=False)
else:
    results = pd.read_csv(RESULT_PATH)
results
```

```python
results.value = results.value.apply(lambda x: round(x, 1))
results.head()
```
